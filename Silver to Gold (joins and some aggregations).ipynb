## One Big Table Model

the next section will further combine the QB stats, draft data and probowl data with the roster and standings dataset. We will also recombine the probowl dataset on the roster and standings dataset to get information of the players on each quarterbacks team.

This will be the base table for our analysis in the next notebook.

```
one_big_table = spark.sql("""
SELECT g.*
, sr.player_name as teammates_name, sr.position as teammate_position, sr.side_off_def as side_of_ball_teammate, sr.games as teammate_games_played, sr.starts as teammate_starts, sr.years_pro as teammate_years_pro, sr.player_value as teammate_player_value
, COALESCE(p.all_pro, false) as teammate_all_pro, COALESCE(p.probowler, false) as teammate_probowler, COALESCE(p.MVP, 0) as teammate_MVP
, sum(sr.player_value) OVER (PARTITION BY g.team, g.year) as roster_player_value
, ss.wins as team_wins, ss.losses as team_loses, ss.alltime_team_wins, ss.alltime_team_loses, ss.rolling_wins, ss.rolling_loses
, ss.ties as team_ties, ss.winning_percentage as year_winning_percentage, ss.division_rank, ss.points_scored, ss.points_allowed, ss.points_differential, ss.strength_of_victory, ss.strength_of_schedule, ss.playoff_seed, ss.playoff_outcome
, (CASE WHEN g.all_pro is TRUE AND g.probowler is not TRUE THEN 
    COALESCE((0.1 * winning_percentage + 0.2 * (quarterback_rating / 122.5) + 0.1 * (total_career_wins / (total_career_wins + total_career_loses)) + 0.05 * total_career_passing_TDs / 581 + 0.05 * total_career_completions / 7127 + 0.05 * total_career_passing_yards / 80137 + 0.05 + 0.05 * fourth_qtr_comebacks / 8 + 0.05 * game_winning_drives / 8 + 0.1 * yards_per_game / 371.2 + 0.1 * adj_yards_per_attempt / 10.5), 0)
    WHEN g.all_pro is TRUE AND g.probowler is TRUE THEN 
    COALESCE(( 0.1 * winning_percentage + 0.2 * (quarterback_rating / 122.5) + 0.1 * (total_career_wins / (total_career_wins + total_career_loses)) + 0.05 * total_career_passing_TDs / 581 + 0.05 * total_career_completions / 7127 + 0.05 * total_career_passing_yards / 80137 + 0.05 + 0.05 + 0.05 * fourth_qtr_comebacks / 8 + 0.05 * game_winning_drives / 8 + 0.1 * yards_per_game / 371.2 + 0.1 * adj_yards_per_attempt / 10.5), 0)
    WHEN g.probowler is TRUE and g.all_pro is not TRUE THEN 
    COALESCE(( 0.1 * winning_percentage + 0.2 * (quarterback_rating / 122.5) + 0.1 * (total_career_wins / (total_career_wins + total_career_loses)) + 0.05 * total_career_passing_TDs / 581 + 0.05 * total_career_completions / 7127 + 0.05 * total_career_passing_yards / 80137 + 0.05 + 0.05 * fourth_qtr_comebacks / 8 + 0.05 * game_winning_drives / 8 + 0.1 * yards_per_game / 371.2 + 0.1 * adj_yards_per_attempt / 10.5), 0)
    else 
    COALESCE(( 0.1 * winning_percentage + 0.2 * (quarterback_rating / 122.5) + 0.1 * (total_career_wins / (total_career_wins + total_career_loses)) + 0.05 * total_career_passing_TDs / 581 + 0.05 * total_career_completions / 7127 + 0.05 * total_career_passing_yards / 80137 + 0.05 * fourth_qtr_comebacks / 8 + 0.05 * game_winning_drives / 8 + 0.1 * yards_per_game / 371.2 + 0.1 * adj_yards_per_attempt / 10.5), 0)
    end) as quarterback_success

, (CASE WHEN g.all_pro is TRUE AND g.probowler is not TRUE THEN 
    COALESCE(( 0.1 * winning_percentage + 0.10 * TDs / 55 + 0.095 * (TDs / (TDs + interceptions)) + 0.3 * (quarterback_rating / 122.5) + 0.04 * fourth_qtr_comebacks / 8 + 0.04 * game_winning_drives / 8 + 0.075 * yards_per_game / 371.2 + 0.075 * adj_yards_per_attempt / 10.5 + 0.1), 0)
    WHEN g.all_pro is TRUE AND g.probowler is TRUE THEN 
    COALESCE(( 0.1 * winning_percentage + 0.10 * TDs / 55 + 0.095 * (TDs / (TDs + interceptions)) + 0.3 * (quarterback_rating / 122.5) + 0.04 * fourth_qtr_comebacks / 8 + 0.04 * game_winning_drives / 8 + 0.075 * yards_per_game / 371.2 + 0.075 * adj_yards_per_attempt / 10.5 + 0.1 + 0.1), 0)
    WHEN g.probowler is TRUE and g.all_pro is not TRUE THEN 
    COALESCE((0.1 * winning_percentage + 0.010 * TDs / 55 + 0.095 * (TDs / (TDs + interceptions)) + 0.3 * (quarterback_rating / 122.5) + 0.04 * fourth_qtr_comebacks / 8 + 0.04 * game_winning_drives / 8 + 0.075 * yards_per_game / 371.2 + 0.075 * adj_yards_per_attempt / 10.5 + 0.1), 0)
    else 
    COALESCE(( 0.1 * winning_percentage + 0.10 * TDs / 55 + 0.095 * (TDs / (TDs + interceptions)) + 0.3 * (quarterback_rating / 122.5) + 0.04 * fourth_qtr_comebacks / 8 + 0.04 * game_winning_drives / 8 + 0.075 * yards_per_game / 371.2 + 0.075 * adj_yards_per_attempt / 10.5), 0)
    end) as quarterback_pop_off_year

from gold_qb_draft_stats_probowl g
  LEFT JOIN silver_roster sr ON (g.year = sr.season AND sr.team = g.team)
  LEFT JOIN silver_standings ss ON (ss.team = g.team AND ss.season = g.year)
  LEFT JOIN silver_probowlers p ON (p.player_name = sr.player_name AND p.season = sr.season)
""")
```
