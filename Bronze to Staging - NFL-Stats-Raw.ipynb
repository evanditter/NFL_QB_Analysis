## Schema Design


```
from pyspark.sql.types import StructField, StructType, StringType, IntegerType, DateType, DoubleType, BooleanType
from pyspark.sql import Row
```

```
# Year	Rk	Player	Tm	Age	Pos	G	GS	QBrec	Cmp	Att	Cmp%	Yds	TD	TD%	Int	Int%	1D	Lng	Y/A	AY/A	Y/C	Y/G	Rate	QBR	Sk	Yds	NY/A	ANY/A	Sk%	4QC	GWD
qb_data_schema = StructType([
StructField("year", IntegerType(), True), 
StructField("qb_rank", IntegerType(), True),
StructField("player", StringType(), True), 
StructField("team", StringType(), True), 
StructField("age", IntegerType(), True),
StructField("position", StringType(), True), 
StructField("games_played", IntegerType(), True), 
StructField("games_started", IntegerType(), True), 
StructField("record", StringType(), True), 
StructField("completions", IntegerType(), True), 
StructField("attempts", IntegerType(), True), 
StructField("completion_perc", DoubleType(), True), 
StructField("yards", IntegerType(), True), 
StructField("TDs", IntegerType(), True), 
StructField("TD_perc", DoubleType(), True), 
StructField("interceptions", IntegerType(), True), 
StructField("int_perc", DoubleType(), True), 
StructField("first_down_passes", IntegerType(), True), 
StructField("longest_throw", IntegerType(), True), 
StructField("yards_per_attempt", DoubleType(), True), 
StructField("adj_yards_per_attempt", DoubleType(), True), 
StructField("yards_per_completion", DoubleType(), True), 
StructField("yards_per_game", DoubleType(), True), 
StructField("quarterback_rating", DoubleType(), True), 
StructField("QBR", DoubleType(), True), 
StructField("sacks", IntegerType(), True), 
StructField("yards_from_sacks", IntegerType(), True), 
StructField("net_yards_per_attempt", DoubleType(), True), 
StructField("adj_net_yards_per_attempt", DoubleType(), True), 
StructField("sack_percent", DoubleType(), True), 
StructField("fourth_qtr_comebacks", IntegerType(), True), 
StructField("game_winning_drives", IntegerType(), True)])

qb_df = spark.read.format("csv").option("header","true").\
schema(qb_data_schema).load("dbfs:/nfl_project/raw/QB_2000_2020_Data.csv")

#qb_df.show(3)
```

```
display(qb_df)
```

```
draft_pick_data_schema = StructType([
StructField("season", IntegerType(), True), 
StructField("team", StringType(), True),
StructField("round", IntegerType(), True), 
StructField("pick_total", IntegerType(), True), 
StructField("pro_football_ref_player_id", StringType(), True),
StructField("player_name", StringType(), True), 
StructField("player_id", StringType(), True), 
StructField("offense_or_defense", StringType(), True), 
StructField("category", StringType(), True), 
StructField("position", StringType(), True)])


dp_df_raw = spark.read.format("csv").option("header","true").\
schema(draft_pick_data_schema).load("dbfs:/nfl_project/raw/draft_picks.csv")

dp_df_raw.show(3)
```

```
game_data_schema = StructType([
StructField("game_id", StringType(), True), 
StructField("season", IntegerType(), True),
StructField("game_type", StringType(), True), 
StructField("week", IntegerType(), True), 
StructField("gameday", DateType(), True),
StructField("weekday", StringType(), True), 
StructField("gametime", StringType(), True), 
StructField("away_team", StringType(), True), 
StructField("away_score", IntegerType(), True), 
StructField("home_team", StringType(), True), 
StructField("home_score", IntegerType(), True), 
StructField("home_location", StringType(), True), 
StructField("result_difference_for_home_team", IntegerType(), True), 
StructField("total_points", IntegerType(), True), 
StructField("overtime", IntegerType(), True), 
StructField("old_game_id", StringType(), True), 
StructField("gsis", IntegerType(), True), 
StructField("nfl_detail_id", IntegerType(), True), 
StructField("espn", IntegerType(), True), 
StructField("away_rest_days", IntegerType(), True), 
StructField("home_rest_days", IntegerType(), True), 
StructField("away_moneyline", StringType(), True), 
StructField("home_moneyline", StringType(), True), 
StructField("spread_line", DoubleType(), True), 
StructField("away_spread_odds", StringType(), True), 
StructField("home_spread_odds", StringType(), True), 
StructField("total_line", IntegerType(), True), 
StructField("under_odds", StringType(), True), 
StructField("over_odds", StringType(), True), 
StructField("divisional_game", DoubleType(), True), 
StructField("roof_type", StringType(), True), 
StructField("surface_type", StringType(), True), 
StructField("temp", IntegerType(), True), 
StructField("wind", IntegerType(), True), 
StructField("away_qb_id", IntegerType(), True), 
StructField("home_qb_id", IntegerType(), True), 
StructField("away_qb_name", StringType(), True), 
StructField("home_qb_name", StringType(), True),
StructField("away_coach", StringType(), True), 
StructField("home_coach", StringType(), True), 
StructField("referee", StringType(), True), 
StructField("stadium_id", IntegerType(), True),
StructField("stadium_name", IntegerType(), True)])


games_df_raw = spark.read.format("csv").option("header","true").\
schema(game_data_schema).load("dbfs:/nfl_project/raw/games.csv")

display(games_df_raw)
```

```
roster_data_schema = StructType([
StructField("season", IntegerType(), True), 
StructField("team", StringType(), True),
StructField("player_id", StringType(), True), 
StructField("player_name", StringType(), True), 
StructField("abbrev_name", StringType(), True),
StructField("side_off_def", StringType(), True), 
StructField("category", StringType(), True), 
StructField("position", StringType(), True), 
StructField("games", IntegerType(), True), 
StructField("starts", IntegerType(), True), 
StructField("years_pro", IntegerType(), True), 
StructField("player_value", IntegerType(), True)])


roster_df_raw = spark.read.format("csv").option("header","true").\
schema(roster_data_schema).load("dbfs:/nfl_project/raw/rosters.csv")

display(roster_df_raw)
```

```
standings_data_schema = StructType([
StructField("season", IntegerType(), True), 
StructField("conference", StringType(), True),
StructField("division", StringType(), True), 
StructField("team", StringType(), True), 
StructField("wins", IntegerType(), True),
StructField("losses", IntegerType(), True), 
StructField("ties", IntegerType(), True), 
StructField("winning_percentage", DoubleType(), True), 
StructField("division_rank", IntegerType(), True), 
StructField("points_scored", IntegerType(), True), 
StructField("points_allowed", IntegerType(), True), 
StructField("points_differential", IntegerType(), True),
StructField("strength_of_victory", DoubleType(), True),
StructField("strength_of_schedule", DoubleType(), True),
StructField("playoff_seed", IntegerType(), True),
StructField("playoff_outcome", StringType(), True)])


standings_df_raw = spark.read.format("csv").option("header","true").\
schema(standings_data_schema).load("dbfs:/nfl_project/raw/standings.csv")

display(standings_df_raw)
```

```
probowlers_data_schema = StructType([
StructField("season", IntegerType(), True), 
StructField("position", StringType(), True),
StructField("player_name", StringType(), True), 
StructField("conference", StringType(), True),
StructField("team", StringType(), True), 
StructField("age", IntegerType(), True), 
StructField("years_pro", IntegerType(), True), 
StructField("games_played", IntegerType(), True), 
StructField("starts", IntegerType(), True), 
StructField("passing_completions", IntegerType(), True), 
StructField("passing_attempts", IntegerType(), True), 
StructField("passing_yards", IntegerType(), True), 
StructField("passing_TDs", IntegerType(), True),
StructField("passing_interceptions", IntegerType(), True), 
StructField("rushing_attemps", IntegerType(), True), 
StructField("rushing_yards", IntegerType(), True), 
StructField("rushing_TDs", IntegerType(), True), 
StructField("receptions", IntegerType(), True), 
StructField("receiving_yards", IntegerType(), True), 
StructField("receiving_TDs", IntegerType(), True), 
StructField("tackles", IntegerType(), True), 
StructField("sacks", IntegerType(), True), 
StructField("interceptions", IntegerType(), True), 
StructField("all_pro_desc", StringType(), True), 
StructField("all_pro_bool", BooleanType(), True), 
StructField("MVP", IntegerType(), True)])


probowlers_df = spark.read.format("csv").option("header","true").\
schema(probowlers_data_schema).load("dbfs:/nfl_project/raw/probowlers_2000_2019.csv")

#probowlers_df.show(3)
display(probowlers_df)
```

## Write to Staging

Now that the files have been loaded into dataframes, we want to write them to disk. In the next notebook, we will cleanse the files to get them ready for analysis. This will include any aggregations or modifications being made to the data. Some initial cleansing was done to the two datasets on QB data an probowl data before enter. 

```
dbutils.fs.rm("nfl_project/staging/qb_stats.parquet", True)
dbutils.fs.rm("nfl_project/staging/draft_picks.parquet", True)
dbutils.fs.rm("nfl_project/staging/games_data.parquet", True)
dbutils.fs.rm("nfl_project/staging/rosters.parquet", True)
dbutils.fs.rm("nfl_project/staging/standings.parquet", True)
dbutils.fs.rm("nfl_project/staging/probowlers.parquet", True)
```

```
%fs
mkdirs nfl_project/staging
```

```
qb_df.select("*").write.save("nfl_project/staging/qb_stats.parquet", format="parquet")
dp_df_raw.select("*").write.save("nfl_project/staging/draft_picks.parquet", format="parquet")
games_df_raw.select("*").write.save("nfl_project/staging/games_data.parquet", format="parquet")
roster_df_raw.select("*").write.save("nfl_project/staging/rosters.parquet", format="parquet")
standings_df_raw.select("*").write.save("nfl_project/staging/standings.parquet", format="parquet")
probowlers_df.select("*").write.save("nfl_project/staging/probowlers.parquet", format="parquet")
```
