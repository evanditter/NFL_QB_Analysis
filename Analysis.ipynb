## Exploratory Analysis

In this notebook, I will explore how a QBs career progresses. This will include looking at individual statistics and accomplishments as well as team (wins/loses and playoffs). I will also try to compare if a higher draft pick is more important or if being drafted to a better team (determined by team record) is more important to the long term success of the team. 

```
winning_view = spark.sql("""select distinct
player, year, quarterback_success, quarterback_pop_off_year, team, round_drafted, pick_total, quarterback_rating, record, wins, loses, total_career_wins, total_career_loses, alltime_team_wins, alltime_team_loses, alltime_team_wins/(alltime_team_loses+alltime_team_wins) as all_time_winning_percentage, rolling_wins, rolling_loses, rolling_wins/(rolling_loses+rolling_wins) as rolling_winning_percentage, year_winning_percentage, playoff_outcome, quarterback_rating, all_pro, probowler, MVP, playoff_seed, playoff_outcome
from gold_one_big_table 
order by team, year, player """)
winning_view.createOrReplaceTempView("winning_view")
```

```
display(winning_view)
```

```
roster_strength = spark.sql("""select distinct
player, year, quarterback_success, quarterback_pop_off_year, team, round_drafted, pick_total, quarterback_rating, record, wins, loses, total_career_wins, total_career_loses,  alltime_team_wins/(alltime_team_loses+alltime_team_wins) as all_time_winning_percentage, rolling_wins/(rolling_loses+rolling_wins) as rolling_winning_percentage, all_pro, probowler, MVP
, COUNT_IF(teammate_all_pro) OVER (PARTITION BY team, year) as number_all_pros
, COUNT_IF(teammate_probowler) OVER (PARTITION BY team, year) as number_probowlers
, SUM(teammate_MVP) OVER (PARTITION BY team, year) as teammate_MVP
, playoff_seed, playoff_outcome
from gold_one_big_table 
where year > 2005
order by team, year, player """)
roster_strength.createOrReplaceTempView("roster_strength_view")
```
